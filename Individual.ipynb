```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Python Assignment - (Input) Rigorbuilder RAW.csv')
df = df.iloc[:,[1,2,3,4]]
```

```python
df['result'] = df['total_statements']+df['total_reasons']
df['newname'] = df['name'].str.upper()
df = df.sort_values(by=['result','newname'],ascending=[False,True])
del df['newname'],df['result']
Rank = np.arange(1,len(df)+1)
df.insert(loc=0,column='Rank',value = Rank)
mapping = {df.columns[1]:'Name',df.columns[2]:'UID',df.columns[3]:'No. of Statements',df.columns[4]:'No. of Reasons'}
df= df.rename(columns=mapping)
df.index = np.arange(1,len(df)+1)
df.set_index('Rank',inplace=True)
df
```

```python
df.to_excel('Individual Leaderboard.xlsx')
```
